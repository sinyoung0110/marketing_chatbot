```python
from langchain_openai import ChatOpenAI 
from dotenv import load_dotenv 
import os

load_dotenv() 
OPEN_API_KEY = os.getenv('OPENAI_API_KEY')

from langchain_core.caches import BaseCache 
from langchain_core.callbacks.base import Callbacks

ChatOpenAI.model_rebuild()

model = ChatOpenAI(model = "gpt-4o-mini", api_key = OPEN_API_KEY) 
model.invoke("최근 로제가 발표한 신곡은 무엇인가요?")
```

```python
from langchain_community.tools import DuckDuckGoSearchResults

search=DuckDuckGoSearchResults(results_separator=';\n')
docs= search.invoke("최근 로제가 발표한 신곡은 무엇인가요?")

print(docs)
```

```python

from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder 

question_answering_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system", 
            "아래 context에 기반하여 사용자의 질문에 답변하라. :\n\n{context}",

        ),
        MessagesPlaceholder(variable_name = "messages"),
    ]
)


document_chain = question_answering_prompt | model
```

```python
from langchain.memory import ChatMessageHistory 

#채팅 메시지를 저장할 메모리 객체 생성
chat_history = ChatMessageHistory() 

#사용자 질문을 메모리에 저장
chat_history.add_user_message("요즘 로제가 발표한 신곡은 무엇인가요?") 

#문서 검색하고 답변 생성
answer = document_chain.invoke( 
    {
        "messages": chat_history.messages,
        "context": docs,
    }
)
#생성된 답변을 메모리에 저장
chat_history.add_ai_message(answer) 

print(answer)
```

```python
#DuckDuckGo API 래퍼를 사용하여 검색할 때 검색 매개변수를 설정하는 클래스 import 
from langchain_community.utilities import DuckDuckGoSearchAPIWrapper

#한국 지역 ("kr-kr")을 기준으로 최근 일주일("w") 내의 검색 결과를 가져오도록 초기화
wrapper=DuckDuckGoSearchAPIWrapper(region="kr-kr", time="w")
```

```python
#검색 기능을 위한 DuckDuckGoSearchResults 초기화 
search = DuckDuckGoSearchResults(
    api_wrapper = wrapper, #앞에서 정의한 api wrapper 등록 
    source = "news", #뉴스로부터만 검색하도록 지정 
    results_separator = ";\n" #결과 항목 사이에 세미콜론과 줄바꿈 구분자 사용
)
```

```python
#로제의 신곡에 대한 반응을 검색하고, 검색 결과를 docs에 저장
docs=search.invoke("로제 신곡 반응")

# 검색 결과 출력
print(docs)
```

```python
#DuckDuckGo를 이용해 ytn.co.kr 웹사이트에서 케이팝데몬헌터스 노래에 대한 분석을 검색함 
docs = search.invoke("site:ytn.co.kr 수능") 
docs
#특정 사이트에서만 검색하도록 지정할 수 있음
#단, 최근 일주일 내의 정보만 가져오므로 최신 뉴스 키워드를 검색해야 함. 
```

```python
#검색 결과의 링크들을 저장할 빈 리스트 초기화
links = []
#검색 결과를 세미콜론과 줄 바꿈 기준으로 분리하고, 각 결과 항목에서 링크를 추출함
for doc in docs.split(";\n"):
    print(doc) #각 검색 결과를 출력하여 확인함 
    link = doc.split("link:")[1].strip() #각 항목에서 url만 추출 
    links .append(link) #추출한 링크를 리스트에 추가
    
#모든 링크를 출력
print(links)
```

```python
#랭체인의 WebBaseLoader를 사용하여 웹 페이지의 내용 불러오기 
from langchain_community.document_loaders import WebBaseLoader
#Webbaseloader 객체 생성 
#(참고) _links'는 웹페이지의 URL 목록을 담고 있는 변수임 
loader = WebBaseLoader(
    web_paths = links, #웹페이지의 링크 목록을 지정함 
    bs_get_text_kwargs={
        "strip":True #웹페이지의 텍스트를 가져올 때 앞뒤의 공백을 제거함
    },
)
#비동기로 웹페이지의 내용을 로드하고, 각 문서를 page_contents 리스트에 추가 
page_contents= [] #각 웹페이지의 내용을 저장할 리스트 
async for doc in loader.alazy_load():
    page_contents.append(doc)#불러온 문서를 page_contents 리스트에 추가

#page_contents에 있는 각 웹페이지의 내용 출력 
for content in page_contents:
    print(content)
    print('--------------')
```

```python
import requests 
from bs4 import BeautifulSoup
#주어진 URL에서 기사 텍스트를 가져오는 함수 
def get_article_text(url):
    try:
        #URL에 get 요청을 보냄 
        response = requests.get(url) 
        
        #요청 실패시 예외 발생 
        response.raise_for_status()
        
        #BeautifulSoup을 사용해 HTML 내용 파싱 
        soup = BeautifulSoup(response.content, 'html.parser')
        
        #기사가 담겨있는 곳의 태그 찾기
        article = soup.find("meta",attrs={"property": "og:description"})
        #기사를 찾았으면 텍스트를 반환함
        if article:
            return article.get("content", ""). strip()
        else:
            try:
                if soup.find('article'):
                    return soup.find('article').get_text(strip=True)
                elif soup.find('div', id="CmAdContent"):
                    return soup.find('div', id="CmAdContent").get_test(strip=True)
            except:
                return "기사 내용을 찾을 수 없습니다."

    #요청 실패시 예외 처리
    except requests.exceptions.RequestException as e:
        return f"URL을 가져오는 중 오류 발생: {e}"
```

```python
#URL 목록이 각 링크를 반복하면서 기사 텍스트 툴력
articles=[] #가져온 내용을 리스트에 담을 것
for link in links:
    print(f"URL: {link}\n")
    article_text = get_article_text(link) 
    print(f"Content:\n{article_text}") 
    print("..................") 
    articles.append(article_text)
```

```python
#검색된 뉴스 본문을 바탕으로 답변을 생성할 것
chat_history.add_user_message("트럼프 대통령과 대한민국")

#문서를 검색해 답변 생성
answer=document_chain.invoke(
    {
        "messages":chat_history.messages,
        "context":docs,
    }
)
# 특정 웹사이트로 검색 범위를 한정하면 언어 모델에게 넘겨주는 텍스트를 불필요한 내용 없이 깔끔하게 정리할 수 있음.
# 이럴 떄 BeautifulSoup 사용 가능

# 하지만, 여러 웹페이지로부터 검색할 경우에는 BeautifulSoup로 전처리하지 못함.-> 토큰을 많이 사용하는 문제 발생하므로 유의하여야 함.
```
