```python
#TavilySearchResults 설치하기(랭체인에서 제공함)
from langchain_community.tools import TavilySearchResults
import os
from dotenv import load_dotenv

load_dotenv()
api_key=os.getenv("TAVILY_API_KEY")

tavily_search=TavilySearchResults(
    max_results=5,
    tavily_api_key=api_key
)
```

```python
res=tavily_search.invoke({"query":'2026년 한국 경제 전망'})
for r in res:
    print(r)
```

```python
#가독성 좋게 출력하기
from IPython.display import JSON
JSON(res).data
```

```python
#langchain에서 제공하는 TavilySearchResult가 아닌 타빌리에서 제공하는 TavilyClient를 사용하면, 해당 웹 페이지의 전체 텍스트까지 불러올 수 있음.

# 웹페이지 전문 불러오기
from tavily import TavilyClient

client=TavilyClient()

content=client.search(
    "2026년 한국 경제 전망",
    search_depth="advanced",
    include_raw_content=True,
)
JSON(content).data
```

```python
# 인터넷에서 자료 조사 후 기사 쓰는 기사 만들기

# 질문에 대한 자료 검색하기
from langchain.adapters.openai import convert_openai_messages
from langchain_openai import ChatOpenAI
from langchain_core.caches import BaseCache 
from langchain_core.callbacks.base import Callbacks

ChatOpenAI.model_rebuild()
llm = ChatOpenAI (model="gpt-4o-mini")

#질문에 대한 자료를 TavilyClinet로 가져오기
#쓰고싶은 기사 내용을 query로 설정하고, 검색한 결과의 웹페이지 내용까지 가져오도록 include_raw_content True로 둠 
query = "현대자동차 미국 시장 2026년 전망"
content = client.search(
    query,
    include_raw_content = True,
    search_depth = "advanced"
)["results"]
```

```python
# 시스템 메시지로 역할 설정

# 프롬프트 설정 
prompt=[{
    "role":"system",
    "content":(
        "당신은 신문 기사를 쓰는 기자 AI 입니다.\n"
        "당신은 주어진 정보를 바탕으로 객관적이고 체계적으로 작성된 기사를 써야 합니다.\n"
    )
},{
    "role":"user",
    "content":(
        f'정보: """{content}"""\n\n'
        f'위의 정보를 사용하여, 다음 질문에 대해 자세한 보고서를 한국어로 작성하세요.: "{query}"\n'
        '-신문 기사 형식을 사용하되, MLA 표준을 준수하는 markdown 문법을 사용해주세요.'
        '-활용한 자료는 출처를 명시하세요.'
    )
}]
```

```python
# 오픈 AI를 랭체인으로 실행하고 출력
report=llm.invoke(prompt).content
print(report)
```
